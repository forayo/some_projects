```python
import pandas as pd
import numpy as np

data=pd.read_csv('samgov.csv',encoding='gbk')

f=open('lines.txt','w')
for i in data['content']:
    f.write(i+'\n')
f.close
```

```python
#Load the stopWords
#do the Word Segmentation and then flitter the stopWords

import jieba
import jieba.posseg as psg

# jieba.load_userdict('userdict.txt')
# create the stopwords-list
def stopwordslist(filepath):
    stopwords = [line.strip() for line in open(filepath, 'r', encoding='utf-8').readlines()]
    return stopwords

def isAllZh(s):
    for c in s:
        if not('\u4e00' <= c <= '\u9fa5'):
            return False
    return True

# To do the Word Segmentation
def seg_sentence(sentence):

    sentence_seged = jieba.cut(sentence.strip())
    stopwords = stopwordslist('stopWords/1893（utf8）.txt')  # load the stopWords
    outstr = ''
    for word in sentence_seged:
        if word not in stopwords:
            if word != '\t' and (word.isalpha() or isAllZh(word)):
                outstr += word
                outstr += " "
    return outstr        

inputs = open('lines.txt', 'r', encoding='utf-8')
outputs = open('lines_seg.txt', 'w',encoding='utf-8')
for line in inputs:
    line_seg = seg_sentence(line)  # the return-value is string
    outputs.write(line_seg + '\n')
outputs.close()
inputs.close()
```

```python
#train LDA model 

import codecs
import gensim
from gensim import corpora
from gensim.models import LdaModel
from gensim.corpora import Dictionary
train = []
# stopwords = codecs.open('stopWords/1893（utf8）.txt','r',encoding='utf8').readlines()
# stopwords = [ w.strip() for w in stopwords ]
fp = open('lines_seg.txt','r')
for line in fp:
    line = line.split()
    #print(line)
    train.append([ w for w in line  ])
print(len(train))
#print(train)
dictionary = corpora.Dictionary(train)
dictionary.save('model_lda_10/lda_10.dict')
corpus = [ dictionary.doc2bow(text) for text in train ]
corpora.MmCorpus.serialize('model_lda_10/lda_10.mm', corpus)
#tfidf = gensim.models.TfidfModel(corpus)
#corpus_tfidf = tfidf[corpus]

lda = LdaModel(corpus=corpus, id2word=dictionary, num_topics=10,passes=20)
#corpus_lda = lda[corpus_tfidf]
#lda.save('model/lda_20.model')
#lda = LdaModel.load('zhwiki_lda.model')
f=open('model_lda_10/lda_10.txt','w',encoding='utf8')
for i in lda.print_topics(10):
    f.write(str(i[0])+' '+str(i[1])+'\n')
f.close()
fp.close()
```

```python
#word frequency count

%matplotlib inline
from jieba.analyse import extract_tags
import numpy as np

f=open('lines_seg.txt','r')
alldata=' '.join([i.strip() for i in f.readlines()])
#extract_tags extract the top 20 frequent word and save its all in tags
tags = extract_tags(sentence=alldata, topK=20)
#jieba.cut，count the frequence of 20 words, Seperately. save as the dict words_freq
words = [word for word in jieba.cut(alldata, cut_all=True)]  
words_freq = {}  
for tag in tags:
    freq = words.count(tag)  
    words_freq[tag] = freq
#sort by the frequence
usedata=sorted(words_freq.items(), key=lambda d:d[1])
tmp= np.array(usedata).T
print(tmp)

from numpy import arange
import matplotlib.pyplot as plt

fig,ax = plt.subplots(figsize=(10,10))
myfont = matplotlib.font_manager.FontProperties(fname='simhei.ttf')

plt.title(u'十九大报告词频统计',fontproperties=myfont,fontsize=20,x=0.001,y=1.02)

ax.set_xlabel(u'出现次数',fontproperties=myfont,fontsize=20,x=0.06,y=1.02,color="gray")

ax.spines['bottom'].set_color('grey')
ax.spines['left'].set_color('grey')
ax.spines['top'].set_color('white')
ax.spines['right'].set_color('white')

tick_positions = range(1,21)
ax.set_yticks(tick_positions)
ax.set_yticklabels(tmp[0],fontproperties=myfont,fontsize=18,color="gray")
bar_positions = arange(20) + 0.75
ax.barh(bar_positions, tmp[1], 0.5,align="edge")
plt.show()
```
